```python
import pyleoclim as pyleo
```

```python
scal = pyleo.utils.jsonutils.json_to_Scalogram('./data/scal_benthic_wwz.json')
```

```python
ts_benthic = scal.timeseries
```

```python
scal_benthic_wwz=ts_benthic.detrend().standardize().wavelet(method='wwz').signif_test(number=5, export_scal=True)
```

```python
spectral = ts_benthic.spectral(method='wwz',scalogram=scal).signif_test(number=5,signif_scals=scal_benthic_wwz.signif_scals).plot()
```
